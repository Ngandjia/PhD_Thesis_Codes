```python
from sage.rings.invariants.invariant_theory import AlgebraicForm, transvectant
```

```python
# V(1)

R.<a0,a1,x,y> = QQ[]                                    

f = AlgebraicForm(2, 1, a0*x + a1*y, x, y)

transvectant(f, f, 1, scale='default') 

# Covariants are generated by f. Invariants are constants.
```

```python
# V(1) + V(1)

R.<a1,b1,a2,b2,x,y> = QQ[]                                    

f1 = AlgebraicForm(2, 1, a1*x + b1*y, x, y)
f2 = AlgebraicForm(2, 1, a2*x + b2*y, x, y)

transvectant(f1, f2, 1, scale='default') 

# Invariants are generated by the above invariant.
```

```python
# V(2)

R.<a1,a2,a3,x,y> = QQ[]                                    

f = AlgebraicForm(2, 2, a1*x^2 + a2*x*y + a3*y^2, x, y)

transvectant(f, f, 2, scale='default')

# Covariants are generated by f and (f,f,2). Invariants are generated by the discriminant (f,f,2).
```

```python
# V(2) + V(2)

R.<a1,a2,a3,b1,b2,b3,x,y> = QQ[]                                    

f1 = AlgebraicForm(2, 2, a1*x^2 + a2*x*y + a3*y^2, x, y)
f2 = AlgebraicForm(2, 2, b1*x^2 + b2*x*y + b3*y^2, x, y)

transvectant(f1, f2, 2, scale='default')
transvectant(transvectant(f1, f2, 1, scale='default'), transvectant(f1, f2, 1, scale='default'), 2, scale='default')
transvectant(f1, f1, 2, scale='default')
transvectant(f2, f2, 2, scale='default')

# Invariants are generated by the above transvectant.
```

```python
# V(3) 

R.<a1,a2,a3,a4,x,y> = QQ[]                                    

f = AlgebraicForm(2, 3, a1*x^3 + a2*x^2*y + a3*x*y^2 + a4*y^3, x, y)

transvectant(transvectant(f, f, 2, scale='default'), transvectant(f, f, 2, scale='default'), 2, scale='default') 

# Invariants are generated by the above transvectant.
```

```python
# V(3)+V(3)

R.<a1,a2,a3,a4,b1,b2,b3,b4,x,y> = QQ[]                                    

f1 = AlgebraicForm(2, 3, a1*x^3 + a2*x^2*y + a3*x*y^2 + a4*y^3, x, y)
f2 = AlgebraicForm(2, 3, b1*x^3 + b2*x^2*y + b3*x*y^2 + b4*y^3, x, y)

transvectant(transvectant(f, f, 2, scale='default'), transvectant(f, f, 2, scale='default'), 2, scale='default') 
transvectant(f1,f2,3)
# Invariants are generated by the above transvectant.
```

```python
# V(4)

R.<a1,a2,a3,a4,a5,b1,b2,b3,b4,b5,x,y> = QQ[]                                    

f1 = AlgebraicForm(2, 4, a1*x^4 + a2*x^3*y + a3*x^2*y^2 + a4*x*y^3 + a5*y^4, x, y)
f2 = AlgebraicForm(2, 4, b1*x^4 + b2*x^3*y + b3*x^2*y^2 + b4*x*y^3 + b5*y^4, x, y)

j21 = transvectant(f1,f1,4);
j22 = transvectant(f2,f2,4);
j23 = transvectant(f1,f2,4);
j31 = transvectant(f1,transvectant(f1,f1,2),4); 
j32 = transvectant(f2,transvectant(f2,f2,2),4); 
j33 = transvectant(f1,transvectant(f1,f2,2),4); 
j34 = transvectant(f2,transvectant(f1,f2,2),4); 
j4 = transvectant(transvectant(f1,f1,2),transvectant(f2,f2,2),4);

j21.polynomial()*6
# Covariants are generated by 5 elements. Two invariants of degree 2 (f,f,4) and 3 (above).
```
